## Scope and KPI

The financial industry has many various metrics for tracking performance of a strategy and portfolio. I would like to go through a few of those as well as take a look at some other specific characteristics.

For this exploration, it would be great to look for a convex strategy with a large number of instances. This would allow for high confidence as well as have the potential to be made better via a closer look at risk management.

The ideal distribution would have the following characteristics:
* Negative tail < Positive tail
    * so the downside is platykurtic while the upside is leptokurtic
* Downside deviation < Upside deviation
* Fatter positive tail, positive skew

Some standard trading metrics:
* Sharpe Ratio
* Net Profit
* Median and Average Profit
* % Winning and % Losing Trades
* Largest Win and Loss
* Average Win and Loss
* Holding Time
* Max Drawdown
* Worst Losing Streak
